### What is the studio with the highest rating?
##### first we have to find the number of productions per studio

```python
studioes=anime['Studio'].value_counts()
studioes
```

```python
# filter and melt the dataframe
m = anime.filter(regex=r'Rating|Studio').melt('Rating', value_name='Studio')

# group and aggregate
add = {'Value_Count': ('Studio', 'count'), 'Average_Rating': ('Rating', 'mean')}
best_studio= m.groupby('Studio', as_index=False).agg(**add)
```

```python
best_studio.sort_values('Average_Rating', ascending=False)
```

#### I should have seen that coming. Less anime would of course have higher ratings. Lets limit it more

```python
top_contributors=best_studio.sort_values('Value_Count', ascending=False).head(25)
top_contributors
```

```python
top_contributors.sort_values('Average_Rating', ascending=False)
```
